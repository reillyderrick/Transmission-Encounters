```python
import pandas as pd

df = pd.read_csv('pio10_results.csv')

df = df.sort_values(by=['dist'])
print(df)
df.to_csv('pio10_results.csv')
```

```python
sig_df = df
sig_df['reach_time_years'] = df['dist']*3.26
sig_df['reach_time_days'] = sig_df['reach_time_years'] * 365.24
sig_df['year'] = sig_df['date'].str[:4]
sig_df['reach_year'] = sig_df['year'].astype(int) + sig_df['reach_time_years'] #approximate reach year
sig_df['return_year'] = sig_df['reach_year'] + sig_df['reach_time_years']

sig_df = sig_df.sort_values(by=['reach_year'])
print(sig_df)
sig_df.to_csv('pio10_results.csv')
```
